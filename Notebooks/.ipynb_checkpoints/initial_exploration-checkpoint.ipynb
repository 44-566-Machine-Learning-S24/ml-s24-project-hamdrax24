```python
import pandas as pd

df = pd.read_csv("DataSets/weatherAUSProcessed.csv")

df.info()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(df[["RainTomorrow", "Pressure3pm"]])
scatter_matrix(df[["RainTomorrow", "Temp3pm"]])
scatter_matrix(df[["RainTomorrow", "WindSpeed3pm"]])
scatter_matrix(df[["RainTomorrow", "Humidity3pm"]])
scatter_matrix(df[["RainTomorrow", "WindDir3pm"]])
```

## First glance observation
At first glance I cannot decifer much from the relationship between each feature and rain tomorrow. With rain tomorrow being only a 1 and 0, every value maps to the 2 extremes making finding a pattern difficult.

```python
scatter_matrix(df[["Humidity3pm", "WindSpeed3pm"]])
scatter_matrix(df[["WindDir3pm", "WindSpeed3pm"]])
scatter_matrix(df[["Temp3pm", "Pressure3pm"]])
scatter_matrix(df[["WindSpeed3pm", "Pressure3pm"]])
```

## Inter feature observation
Here we notice some correlation between the features such as:
 * Higher pressure tends to result in lower wind speeds.
 * Certain directions have noticable higher max wind speeds.
 * Pressure and temprature increase in correlation as temprature increases.
